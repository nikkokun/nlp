```python
def unicode_normalize(cls, s):
    pt = re.compile('([{}]+)'.format(cls))

    def norm(c):
        return unicodedata.normalize('NFKC', c) if pt.match(c) else c

    s = ''.join(norm(x) for x in re.split(pt, s))
    s = re.sub('－', '-', s)
    return s

def remove_extra_spaces(s):
    s = re.sub('[ 　]+', ' ', s)
    blocks = ''.join(('\u4E00-\u9FFF',  # CJK UNIFIED IDEOGRAPHS
                      '\u3040-\u309F',  # HIRAGANA
                      '\u30A0-\u30FF',  # KATAKANA
                      '\u3000-\u303F',  # CJK SYMBOLS AND PUNCTUATION
                      '\uFF00-\uFFEF'   # HALFWIDTH AND FULLWIDTH FORMS
                      ))
    basic_latin = '\u0000-\u007F'

    def remove_space_between(cls1, cls2, s):
        p = re.compile('([{}]) ([{}])'.format(cls1, cls2))
        while p.search(s):
            s = p.sub(r'\1\2', s)
        return s

    s = remove_space_between(blocks, blocks, s)
    s = remove_space_between(blocks, basic_latin, s)
    s = remove_space_between(basic_latin, blocks, s)
    return s

def normalize_neologd(s):
    s = s.strip()
    s = unicode_normalize('０-９Ａ-Ｚａ-ｚ｡-ﾟ', s)

    def maketrans(f, t):
        return {ord(x): ord(y) for x, y in zip(f, t)}

    s = re.sub('[˗֊‐‑‒–⁃⁻₋−]+', '-', s)  # normalize hyphens
    s = re.sub('[﹣－ｰ—―─━ー]+', 'ー', s)  # normalize choonpus
    s = re.sub('[~∼∾〜〰～]', '', s)  # remove tildes
    s = re.sub('[0-9]', '', s)
    s = s.translate(maketrans('!"#$%&\'()*+,-./:;<=>?@[¥]^_`{|}~｡､･｢｣','！”＃＄％＆’（）＊＋，－．／：；＜＝＞？＠［￥］＾＿｀｛｜｝〜。、・「」'))
    s = re.sub(re.compile("[!-/:-@[-`{-~]"), '', s)
    s = re.sub(re.compile('[!"#$%&\'()*+,-./:;<=>?@[¥]^_`{|}~｡､･｢｣]'), '', s)
    s = re.sub(re.compile('[■□◆◇◯“…【】『』！”＃＄％＆’（）＊＋，－．／：；＜＝＞？＠［￥］＾＿｀｛｜｝〜。、・「」]'), '', s)

    s = remove_extra_spaces(s)
    s = unicode_normalize('！”＃＄％＆’（）＊＋，－．／：；＜＞？＠［￥］＾＿｀｛｜｝〜', s)  # keep ＝,・,「,」
    s = re.sub('[’]', '\'', s)
    s = re.sub('[”]', '"', s)
    s = s.replace('\n','')
    s = s.replace('\r','')
    return s


def tokenize(text):
    node = mecab.parseToNode(text)
    while node:
        if node.feature.split(',')[0] in ['名詞', '動詞', '形容詞']:
            yield node.surface.lower()
        node = node.next

def check_stopwords(word):
    return False

def get_words_main(content):
    return [token for token in tokenize(content) if not check_stopwords(token)]

def preprocess(content):
    return get_words_main(normalize_neologd(content))
```

```python
from urllib import request
from gensim.models.doc2vec import Doc2Vec
from gensim.models.doc2vec import TaggedDocument

#辞書ファイルの指定
mecab = MeCab.Tagger('-M4 -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')
mecab.parse('')
```

```python
# 新しく学習したいときだけ実行

res = request.urlopen("http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt")
stopwords = [line.decode("utf-8").strip() for line in res]
res = request.urlopen("http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/English.txt")
stopwords += [line.decode("utf-8").strip() for line in res]


contents = df_training.apply(lambda x: '{} {}'.format(x[0], x[1]), axis=1).tolist()
cleaned_contents = [normalize_neologd(text) for text in contents]
words = [get_words_main(text) for text in cleaned_contents]

training_docs = []
for i, w in enumerate(words):
    w = [x for x in words[i] if x not in stopwords]
    sent = TaggedDocument(words=w, tags=[df_training.iloc[i].title])
    training_docs.append(sent)

model = Doc2Vec(documents=training_docs, size=300, min_count=5)
model.save('doc2vec_20180710_1305.model')
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def similar_doc_mix(docs):
    tmp_vec = np.array([0.0]*300)
    for i in range(len(docs)):
        tmp_vec += model.infer_vector([x for x in get_words_main(normalize_neologd(docs[i])) if x not in stopwords])
    tmp_vec /= len(docs)
    #print(tmp_vec)
    similar_doc_20 = model.docvecs.most_similar([tmp_vec], topn=10)
    return similar_doc_20


recommendation_source = doc_vector_list
recommendation_docs = df_test

# 学習に使ったのとは別のドキュメントから、ページをおすすめ
def get_similar_doc_by(docs):
    tmp_vec = np.array([0.0]*300)
    for i in range(len(docs)):
        model.infer_vector(preprocess(docs[i]))
        tmp_vec += model.infer_vector([x for x in preprocess(docs[i])])
    recommended_docs = []
    for index, doc_vec in enumerate(recommendation_source):
        cosine_similarity_array = cosine_similarity([tmp_vec], [doc_vec])
        cosine_similarity_val = cosine_similarity_array[0][0]
        #print(cosine_similarity_val)
        if cosine_similarity_val >= 0.4:
            #print(recommendation_docs.iloc[index])
            #print(str(index) + "," + str(cosine_similarity_val) + (recommendation_docs[index]['title'] + recommendation_docs[index]['body'])[0:100])
            recommended_docs.append({
                "similarity" : round(cosine_similarity_val,3),
                "title" : recommendation_docs.iloc[index]['title'],
                "body" : recommendation_docs.iloc[index]['body'],
                "index" : index
            })
    return sorted(recommended_docs, key=lambda x: -x["similarity"])[:10]
```
